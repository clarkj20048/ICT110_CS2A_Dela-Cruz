## CHALLENGE


### CHALLENGE 1

```python
total = 0
count = 0

while True:
  line = input("Enter a number: ")
  if line == 'done':
    break
  try:
    number = float(line)
    total = total + number
    count = count + 1
  except:
    print("Invalid input")
    continue

if count == 0:
  print("No numbers entered.")
else:
  average = total / count
  print(total, count, average)
```

### CHALLENGE 2

```python
max = None
min = None
count = 0

while True:
  line = input("Enter a number: ")
  if line == 'done':
    break
  try:
    number = int(line)
    if count == 0:
      max = number
      min = number
    else:
      for num in [number, max, min]:
        if num > max:
          max = num
        if num < min:
          min = num
    count += 1
  except:
    print("Invalid input")
    continue

if count == 0:
  print("No numbers entered.")
else:
  print("Maximum:", max)
  print("Minimum:", min)
```
