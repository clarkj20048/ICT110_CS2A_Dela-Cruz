#### Exercise 1

Given the code below    

    text = 'Business Analytics'

What will the following expressions produce?
    
    text[4]
    text[2:4]
    text[:4]
    text[4:]
    text[-2]
    text[-4:-2]
    text[-2:]
    text[:-2]
    

Answer :

```python
text = 'Business Analytics'

text[4]
print(text[4])      #n

text[2:4]
print(text[2:4])    #si

text[:4]
print(text[:4])     #Busi

text[4:]
print(text[4:])     #ness Analytics

text[-2]
print(text[-2])     #c

text[-4:-2]
print(text[-4:-2])  #ti

text[-2:]
print(text[-2:])    #cs

text[:-2]
print(text[:-2])    #Business Analytics
```

#### Exercise 2

Given the code below:
    
    phrase = 'One Ring to rule them all'

What string slicing expression gives out the following words:

    'One '
    'Ring'
    'all'
    'them'
    'One Ring to rule them all'
    

Answer:

```python
phrase = 'One Ring to rule them all'

print(phrase[:3])     #One
print(phrase[4:8])    #Ring
print(phrase[-3:])    #all
print(phrase[-8:-4])  #them
print(phrase[:])      #One Ring to rule them all
```
