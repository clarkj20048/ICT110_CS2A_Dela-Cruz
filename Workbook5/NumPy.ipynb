```python
#A.

import numpy as np

A = np.array([[i for i in range(1, 11)] for _ in range(10)])

print("A:\n", A)

print("Shape of A:", A.shape)

#np.array(): Creates a new ndarray object.
#[[i for i in range(1, 11)] for _ in range(10)]: Creates a nested Python list containing values from 1 to 100.
#A.shape: Returns the shape of the ndarray object A.
```

```python
#B.

B = np.ones((5, 5), dtype=int)

print("B:\n", B)

print("Shape of B:", B.shape)
print("Size of B:", B.size)
print("Dimensions of B:", B.ndim)

#np.ones(): Creates a new array of given shape and type, filled with ones.
#(5, 5): Specifies the shape of the array to be created.
#dtype=int: Specifies the data type of the array elements as integers.
```

```python
#C.

C = np.arange(-5, 6)

C *= 2

print("C:", C)

#np.arange(-5, 6): Creates an array of values from -5 to +5.
#C *= 2: Multiplies each value in the array C by 2.
#print("C:", C): Prints the resulting array C.
```

##Exercise 3

A. Create a four-dimensional array of random integers in the interval [0, 12]. Print the result.
1. Get the dimensions, shape, and size.
    * Explain what is the result briefly by writing a comment before the code.

B. Get the maximum number for each dimensions and display it.
   * Explain what is the result briefly by writing a comment before the code.

C. Multipy each array with 4. What is its result?
    * Write a brief explanation on the multiplication process.


```python
#A.

#np.random.randint(0, 13, size=(2, 3, 4, 5)): Generates a four-dimensional array of random integers ranging from 0 to 12. The dimensions of the array are specified as (2, 3, 4, 5).

import numpy as np

four_dim_array = np.random.randint(0, 13, size=(2, 3, 4, 5))

print("Four-dimensional Array:\n", four_dim_array)
```

```python
#B.

#four_dim_array.max(axis=(0, 1, 2, 3)): Finds the maximum number along all axes (dimensions) of the four-dimensional array.

max_values = four_dim_array.max(axis=(0, 1, 2, 3))

print("Maximum number for each dimension:", max_values)
```

```python
#C.

four_dim_array *= 4

print("Array after multiplication:\n", four_dim_array)

#four_dim_array *= 4: Multiplies each element in the four-dimensional array by 4. This operation scales the entire array by a factor of 4, increasing all values proportionally.
```
